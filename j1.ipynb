```python
# Finding pi through area of circle
import random
import numpy as np
import matplotlib.pyplot as plt
nc=0
i=0
inx, iny, ox, oy, pi= [],[],[],[], []
for n in range (0,10000):
    while (i<=n):
        x= random.random()
        y=random.random()
        i+=1
        if (((x**2)+(y**2))<1):
            nc+=1
            inx.append(x), iny.append(y)
        else:
            ox.append(x), oy.append(y)
    pi.append(4*nc/i)

plt.plot(pi)
plt.axhline(np.pi, linestyle='--',color='crimson')
plt.grid()
plt.show()
```

```python
plt.scatter(inx, iny, color='teal', alpha=0.6, s=9)
plt.scatter(ox, oy, color='lightpink', alpha=0.6, s=3)
plt.axis("square")
```

```python
# INTEGRATION Method 1
import random
import matplotlib.pyplot as plt
nc=0
inx,iny, ox, oy=[], [] ,[],[]
```

```python
## x^2
for i in range(0,10000):
    x=random.uniform(0,2)
    y=random.uniform(0,4)
    if (y<(x*x)):
        inx.append(x)
        iny.append(y)
        nc+=1
    else:
        ox.append(x)
        oy.append(y)
print("Integral of x^2 from 0 to 2: ",4*2*nc/1000)           # M_0*(b-a)*Nc/Nt
plt.scatter(inx,iny,color='pink',label='points under x^2',s=1)
plt.scatter(ox,oy,color='blue',label='outside points',s=2)
plt.axis("square")
plt.legend()
plt.show()
```

```python
## (1/2)*((5*x**3)-3*x)
from scipy.optimize import minimize

def f(x):
    return (1/2)*((5*x**3)-3*x)
def neg_f(x):
    return -((1/2)*((5*x**3)-3*x))
x0= [0]

val= minimize(neg_f, x0, bounds=[(-1,1)])
m0= -val.fun
print(val.x)
print(m0)
print(f(val.x))
```

```python
for i in range(0,1000000):
    x=random.uniform(-0.75,0)              
    y=random.uniform(0,m0)
    if (y< f(x)):
        inx.append(x)
        iny.append(y)
        nc+=1
    else:
        ox.append(x)
        oy.append(y)                # f_avg= 1/(b-a) * integral f(x) from a to be

print("Integral of P_3(x) from -(0.75) to 0: ", m0*(0.75) *nc/1000000)          # M_0*(b-a)*Nc/Nt
```

```python
plt.scatter(inx,iny,color='pink',label='points under P_3',s=0.8, alpha=0.8)
plt.scatter(ox,oy,color='blue',label='outside points',s=.8,alpha=0.8)
plt.title("Graph of P_3(x)=(1/2)*(5x^3 - 3x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(-0.8,0)
plt.ylim(-0.02,0.48)
plt.grid()
plt.show()  
```

```python
# Integration method 2 (using M.C.)
import numpy as np
import matplotlib.pyplot as plt
f=lambda x: (1/2)*((5*x**3)-3*x)
xl=np.linspace(-0.75,0,10000)

avg=0
for i in xl:               # integral from -1/2 to 0
    avg+=f(i)
avg=avg/10000                # f_avg= 1/(b-a) * integral f(x) from a to b
print("Average = ",avg)
print("Integral of P_3(x) from -(0.75) to 0: ",avg*0.75) 

x=np.linspace(-0.75,1,1000)
y=f(x)
plt.plot(x,y)
plt.title("Graph of P_3(x)=(1/2)*(5x^3 - 3x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()  
```

```python
# Finding Integral (Volume), multivar
import numpy as np
import random
f= lambda x,y: 4-x-y
xl=np.linspace(0,2,1000)
yl=np.linspace(0,1,1000)
nc=0    # no. of darts falling within the required range/curve
# max value
m=0           
for i in range(0,1000):
    if (m<f(xl[i],yl[i])):
        m= f(xl[i],yl[i])
# Throwing Darts
for i in range(0,1000):
    z=random.uniform(0,m)
    x=random.uniform(0,2)
    y=random.uniform(0,1)
    if (z<f(x,y)):
        nc+=1 
print((2*1*m)*nc/1000)    # vol. of f(x,y)/vol. of cuboid = Nc/Nt
```
