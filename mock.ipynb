```python
import numpy as np
import plotly.graph_objects as go

# Create parameter ranges
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)       # grid of (x, y) coordinate pairs
Z = np.sin(X) * np.cos(Y)             # Z[i][j] = sin(X[i][j]) * cos(Y[i][j])


# Plotting
fig1 = go.Figure(data=[go.Surface(z=Z, x=x, y=y)])

# Add contour lines on the Z axis
fig1.update_traces(contours_z=dict(
    show=True,
    usecolormap=True,
    highlightcolor="limegreen",
    project_z=True
))

# Customize camera view and layout
fig1.update_layout(
    title="Sine-Cosine Wave Surface sin(x)*cos(y)",
    autosize=False,
    width=900,
    height=900,
    scene_camera_eye=dict(x=1.6, y=-1, z=1),
    margin=dict(l=60, r=60, b=60, t=60)
)
fig = go.Figure( data=go.Contour(z=Z, x=x, y=y, colorscale='Viridis',contours=dict(coloring='heatmap',showlabels=True)) )
fig.update_traces(
    contours=dict(coloring='heatmap',showlabels=True, labelfont=dict(size=12, color='white')),line_smoothing= 1.3 )
fig.show()
fig1.show()
```

```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np
from scipy.special import iv, kv , gammainc, gamma
from scipy.integrate import quad
import emcee
import multiprocessing, time
import corner
from numba import njit

k=7.6695
G= 4.302e-6

data= pd.read_csv("M31 Grand Rotation Curve_data", delim_whitespace=True, header=None, names=['x', 'y', 'error','no'])
xl=data['x']
yl=data['y']
el=data['error']
xl= np.array(xl)
yl = np.array(yl)
el = np.array(el)

# BULGE
@njit
def sigbr_x(x, sig_b, a_b):
    return sig_b * np.exp(-k * ((x / a_b)**0.25 - 1))

@njit
def dsigbr_x_dx(x, sig_b, a_b):
    prefactor = -(k / 4) * (1 / a_b)**0.25 * x**(-0.75)
    return prefactor * sigbr_x(x, sig_b, a_b)

@njit
def rho_bulge_numba(r, a_b, sig_b):
    result = 0.0
    du = (20.0 - 0.01) / 200
    for i in range(201):
        u = 0.01 + i * du
        x = r * np.cosh(u)
        integrand = dsigbr_x_dx(x, sig_b, a_b)
        weight = 1.0
        if i == 0 or i == 200:
            weight = 0.5
        result += weight * integrand
    integral = result * du
    return -integral / np.pi

@njit
def vb_numba(r, a_b, sig_b):
    if r <= 0:
        return 0.0
    result = 0.0
    dr = (r - 0.01) / 200
    for i in range(201):
        ri = 0.01 + i * dr
        integrand = rho_bulge_numba(ri, a_b, sig_b) * ri**2
        weight = 1.0
        if i == 0 or i == 200:
            weight = 0.5
        result += weight * integrand
    integral = result * dr
    vb2 = (4 * np.pi * G / r) * integral
    if vb2 > 0:
        return np.sqrt(vb2)
    return 0.0

# DISK
def vd_generic(r, a_d, sig_d):
    """
    Compute disk velocity array for given constants sigma_d and a_d.
    Accepts scalar or array r.
    """
    r = np.atleast_1d(r)
    # Avoid r <= 0
    valid = r > 0
    v2 = np.zeros_like(r)
    # Compute only for valid r
    x = r[valid] / (2 * a_d)
    term = iv(0, x) * kv(0, x) - iv(1, x) * kv(1, x)
    v2_val = np.pi * G * sig_d * (r[valid]**2 / a_d) * term
    # Non-negative check
    v2[valid] = np.where(v2_val > 0, v2_val, 0.0)
    # Return velocity
    v = np.sqrt(v2)
    # If scalar input, return scalar
    # if v.ndim == 0:
    #     return float(v)
    if v.size == 1:
        return v.item()
    else:
        return v
    
# HALO
def vh_ein_vec(r, rho_0, h, n):
    r = np.atleast_1d(r)
    v2 = np.zeros_like(r)
    x = (r[ r > 0] / h)**(1/n)
    mass_val = 4 * np.pi * rho_0 * n * h**3 * gamma(3*n) * gammainc(3*n, x)
    v2_val = (G * mass_val) / r[ r > 0]
    v2[ r > 0] = np.where(v2_val > 0, v2_val, 0.0)
    v = np.sqrt(v2)
    return float(v) if v.shape == (1,) else v
```

```python
# xl= xl[:-1]
# yl= yl[:-1]
# el=el[:-1]
def ln_likeli_ein(xl ,yl,el ,a_b, sig_b, a_d, sig_d, rho, h, n):
    # if ([el<=0]):
    #     return -np.inf
    el_safe = np.where(el == 0, np.finfo(float).eps, el)
    vbulge = np.array([vb_numba(x, a_b, sig_b ) for x in xl])
    vdisk  = np.array([vd_generic(x, a_d, sig_d ) for x in xl])
    vhalo  = np.array([vh_ein_vec(x, rho, h, n) for x in xl])
    vth = np.sqrt(vbulge**2 +vdisk**2 + vhalo**2)
    chi2 = np.sum((yl - vth)**2 / el_safe**2)
    return -0.5 * chi2

def ln_prior_ein(theta):
      a_b, sig_b, a_d, sig_d, rho, h, n = theta
      if not (0.5 < a_b <2.0):
          return -np.inf
      if not (0.5e9 < sig_b < 1.5e9):
          return -np.inf
      if not ( 7 < a_d < 12 ):
          return -np.inf
      if not (0.2e9 < sig_d < 0.8e9):
          return -np.inf
      if not (8e7< rho <12e7):
          return -np.inf
      if not ( 2< h< 6):
         return -np.inf
      if not ( 0.2<n< 0.8):
          return -np.inf
      return 0.0

def ln_post_ein(theta, xl, yl, el):
    a_b, sig_b, a_d, sig_d, rho, h, n = theta
    # if any([
    # a_b <= 0, sig_b <= 0,
    # a_d <= 0, sig_d <= 0,
    # rho <= 0, h <= 0, 
    # n <= 0   
    #     ]):                        # if you have n
    #     return -np.inf
    lp = ln_prior_ein(theta)
    ll = ln_likeli_ein(xl, yl, el,a_b, sig_b, a_d, sig_d, rho, h, n)
    post= lp + ll
    if not np.isfinite(post):
        return -np.inf
    return lp + ll

#plot_residuals(xl,yl,vth)
```

```python
init_guess = [1.309, 0.731e9 , 8.990, 0.531e9, 10.755e7, 4.067, 0.430]
for _ in range(20):
    theta = init_guess + np.random.normal(0, [ 0.075, 0.05e9, 0.25, 0.03e9, 0.2e7, 0.2, 0.02])
    t0 = time.time()
    val = ln_post_ein(theta, xl, yl, el)
    print(f"ln_post_ein = {val:.9f}, Time = {time.time() - t0:.4f} sec")
```

```python
# ---- MCMC SETUP ----
ndim = 7          # No. of parameters
nwalkers = 50      # Number of independent chains ("walkers")
nsteps = 5000      # Steps each walker takes
burnin = 500      # Discard these initial steps (they're still "warming up")

# ---- INITIAL GUESS FOR WALKERS ----
np.random.seed(42)
init_guess = [1.309, 0.731e9 , 8.990, 0.531e9, 10.755e7, 4.067, 0.430]  # Rough center of prior a_b, sig_b, a_d, sig_d, rho, h
pos =  np.array([init_guess] * nwalkers) + np.random.normal(0, [0.075, 0.05e9, 0.25, 0.03e9, 0.2e7, 0.2, 0.02 ], (nwalkers, ndim))

if __name__ == '__main__':
    with multiprocessing.Pool() as pool:
        sampler_e = emcee.EnsembleSampler(
            nwalkers,
            ndim,
            ln_post_ein,
            args=(xl, yl, el),
            pool=pool
        )
        sampler_e.run_mcmc(pos, nsteps, progress=True)

    flat_samples_e = sampler_e.get_chain(discard=burnin,  flat=True)      # thin, used to reduce autocorrelation by keeping every thin-th sample
    print("Shape of samples:", flat_samples_e.shape)
```
